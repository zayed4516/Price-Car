```python
categroical=df.select_dtypes(include='object')
categroical.head()
```

```python
general_mean=df['CO2EMISSIONS'].mean()
general_mean
```

```python
for col in categroical:
  global_mean=df.groupby(col)['CO2EMISSIONS'].agg(['mean'])
  graph=sns.barplot(x=col,y='mean',data=global_mean)
  graph.axhline(general_mean,color='red')
  plt.xticks(rotation=90)
  plt.text(0,general_mean +5,"global_mean",color='black')
  plt.xlabel(col)
  plt.ylabel('CO2EMISSIONS')
  plt.show()
```

```python
Data=df[['FUELCONSUMPTION_CITY','CO2EMISSIONS','MAKE','ENGINESIZE']]
```

```python
Data
```

```python
x=Data.drop('CO2EMISSIONS',axis=1)
y=Data['CO2EMISSIONS']
```

```python
x=pd.get_dummies(x,columns=['MAKE'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
sns.regplot(x=y_test,y=y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
```
