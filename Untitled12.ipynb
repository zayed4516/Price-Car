```python
data=pd.get_dummies(df,columns=['Make','Model','Engine Fuel Type','Transmission Type','Driven_Wheels','Market Category','Vehicle Size','Vehicle Style'])
```

```python
data.head()
```

```python
numerical.corr()
```

```python
x=data.drop(['MSRP','Number of Doors','highway MPG','Popularity','city mpg',],axis=1)
y=data['MSRP']
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[0]
```

```python
y_test[0]
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
sns.regplot(x=y_test,y=y_pred,color='red')
plt.xlabel('y_test')
plt.ylabel('y_pred')
```
